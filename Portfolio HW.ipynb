```python
start = dt.datetime(2000,1,1)
end = dt.datetime.now()
```

## I did each portfolio in the tail version instead of head because I like to see what they are doing today.

```python
def get_stock(ticker):
    data = pdr.DataReader(f"{ticker}","yahoo", start, end)
    data[f'{ticker}'] = data["Adj Close"]
    data = data[[f'{ticker}']]
    print(data.tail())
    return data
```

# Here I am building my same industry portfolio.

```python
Target = get_stock('TGT')
Walmart = get_stock('WMT')
Dollar_General = get_stock('DG')
Kroger = get_stock('KR')
Ross = get_stock('ROST')
Academy = get_stock('ASO')
Kohls = get_stock('KSS')
Burlington = get_stock('BURL')
Home_Depot = get_stock('HD')
Lowes = get_stock('LOW')
```

```python
stocks = ['TGT', 'WMT', 'DG', 'KR', 'ROST', 'ASO', 'KSS', 'BURL', 'HD', 'LOW']
```

```python
combine_stocks(stocks)
```

```python
from functools import reduce

def combine_stocks(tickers):
    data_frames = []
    for i in tickers:
        data_frames.append(get_stock(i))
    
    df_merged = reduce(lambda left, right: pd.merge(left, right, on=['Date'], how='outer'),data_frames)
    print(df_merged.tail())
    return df_merged
```

```python
same_ind_stocks = ['TGT', 'WMT', 'DG', 'KR', 'ROST', 'ASO', 'KSS', 'BURL', 'HD', 'LOW']
same_ind_portfolio = combine_stocks(stocks)
```
